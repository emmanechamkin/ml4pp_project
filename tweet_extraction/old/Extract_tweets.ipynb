```python
import tweepy
import creds
import json
import csv
import time
import random
```

```python
token = creds.access_token
secret_token = creds.access_secret_token
key = creds.API_key
secret_key = creds.API_secret_key

auth = tweepy.OAuthHandler(key, secret_key)
auth.set_access_token(token, secret_token)

api = tweepy.API(auth)
```

```python
queries = ["Chicago Police", "CPD", "chicago police department", "second city cop", "chicago cop"]
tweets = []
for q_ in queries:
    sleeper_min = random.randint(15, 25)
    time.sleep(sleeper_min * 60)
    for tweet in tweepy.Cursor(api.search,
                           q=q_,
                           count=100,
                           result_type="recent",
                           include_entities=True,
                           lang="en").items():
        tweets.append(tweet.text)
```

```python
with open("output.csv", "a") as csv_file:
    csv_writer = csv.writer(csv_file)
    for tweet in tweets:
        csv_writer.writerow([tweet])
```
