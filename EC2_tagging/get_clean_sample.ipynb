### Purpose
This was a super quick script that I actually did in ipython and copied here to get a sense of what a random sample would look like. 

```python
import pandas as pd
import numpy as np
original_draw = pd.read_csv("cleaned_coded_april.csv", header=None, names=["tweet", "sentiment"])
```

```python
filtered_tweets = original_draw[original_draw["sentiment"] != "Neutral"]
```

```python
np.random.seed(0)
filtered_tweets[["tweet"]].sample(1000).to_csv("cleaned_dropped_april.csv")
```
